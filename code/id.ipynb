```python
import csv

# Schritt 1: Daten aus der CSV-Datei laden
data = []
with open('/Users/takomon/Documents/IT/TechLabs/team2/team02/datasets/2.landing/ladesaeulen_exc.csv', 'r') as csvfile:
    csvreader = csv.DictReader(csvfile)
    for row in csvreader:
        data.append(row)

# Schritt 2: Eindeutige IDs hinzufügen
for idx, entry in enumerate(data):
    entry['ID'] = idx + 1  # Einfacher Ansatz: Verwende die Index-Position plus 1 als ID

# Reihenfolge der Spaltennamen festlegen (ID zuerst, dann die restlichen Spalten)
fieldnames = ['ID'] + list(data[0].keys())

# Schritt 3: Neue CSV-Datei mit aktualisierten Daten erstellen
output_file = "/Users/takomon/Documents/IT/TechLabs/team2/team02/datasets/2.landing/ladesauele_id.csv"
with open(output_file, 'w', newline='') as csvfile:
    csvwriter = csv.DictWriter(csvfile, fieldnames=fieldnames)
    csvwriter.writeheader()
    csvwriter.writerows(data)
    

print("IDs wurden hinzugefügt, und die aktualisierten Daten wurden in", output_file, "gespeichert.")
```
