```python
df_zulassung_bundesland = pd.read_excel("/Users/philippusasefaw/team02/datasets/fz14_2022.xlsx", sheet_name="FZ 14.1")

df_zulassung_bundesland = df_zulassung_bundesland.drop([0,1,2,3,4,5,7,8])
df_zulassung_bundesland = df_zulassung_bundesland.rename(columns= df_zulassung_bundesland.iloc[0])
df_zulassung_bundesland = df_zulassung_bundesland.drop(df_zulassung_bundesland.columns[[0,5,8,9,10,11,12,13,15]], axis = 1)
df_zulassung_bundesland = df_zulassung_bundesland.reset_index(drop = True)
df_zulassung_bundesland["Land"] = df_zulassung_bundesland.ffill(inplace = True)
df_zulassung_bundesland = df_zulassung_bundesland.drop(df_zulassung_bundesland[df_zulassung_bundesland["Kraftstoffart"]!="Elektro (BEV)"].index, axis=0)
df_zulassung_bundesland = df_zulassung_bundesland.drop(df_zulassung_bundesland.columns[[0,9]], axis = 1)
df_zulassung_bundesland = df_zulassung_bundesland.reset_index(drop = True)
df_zulassung_bundesland.rename(columns={"Kraftomni-\nbusse": "Kraftomnibusse"}, inplace=True)

df_zulassung_bundesland.head(50)
```
